```python
# Function to draw protractor-like annotations
def draw_protractor(ax):
    # Draw the circle
    circle = plt.Circle((0, 0), 1, color="gray", fill=False, linestyle="dotted")
    ax.add_artist(circle)
    
    # Annotate specific angles (0, 30, 45, 60, 90 degrees)
    for angle in [0, 15, 30, 45, 60, 90]:
        angle_rad = np.radians(angle)
        x = np.cos(angle_rad)
        y = np.sin(angle_rad)
        ax.plot([0, x], [0, y], color="gray", linestyle="dotted")
        ax.text(x * 1.1, y * 1.1, f"{angle}Â°", fontsize=10, color="black", ha="center", va="center")
```

# 0-90 degrees

```python
import numpy as np
import matplotlib.pyplot as plt

N_mc = 1000
ALPHAS = []
alphas = []

for i in range(N_mc):
    # Draw a random alpha, merger axis angles with respect to the sky
    ALPHA = np.random.rand() * (np.pi / 2)
    ALPHA_degrees = np.degrees(ALPHA) # radian to degrees
    ALPHAS.append(ALPHA_degrees)
    
    # Since not all alpha are equally likely.
    alpha = (1 - np.cos(ALPHA)) * (np.pi / 2)
    alpha_degrees = np.degrees(alpha)  # radian to degrees
    alphas.append(alpha_degrees)

# プロットの作成
plt.hist(ALPHAS, bins=30, edgecolor='black')
plt.xlabel('ALPHA[degree]')
plt.ylabel('Frequency')
plt.title('Distribution of ALPHA')
plt.show()

# プロットの作成
plt.hist(alphas, bins=30, edgecolor='black')
plt.xlabel('alpha[degree]')
plt.ylabel('Frequency')
plt.title('Distribution of alpha')
plt.show()
```

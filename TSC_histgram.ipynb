```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

## Dawson's Paper

```python
# Define the PDF function
def PDF_TSC(TSC, a, b):
    return 0.5 * (1 - np.tanh((TSC - 0.5 * a) / (0.25 * b)))

# Parameters for the sigmoid function
a = 0.4  # FWHM of the duration of the temperature boost
b = 1.4  # Entire boost duration
TSC_values = np.linspace(-1.5, 3, 1000)  # TSC values from -1.5 to 3 Gyr

# Calculate the PDF values
PDF_values = PDF_TSC(TSC_values, a, b)

# Create a histogram plot
plt.figure(figsize=(10, 6))
plt.plot(TSC_values, PDF_values, label='PDF(TSC)', color='blue')
plt.fill_between(TSC_values, PDF_values, alpha=0.3, color='blue')
plt.title('Probability Density Function of TSC')
plt.xlabel('TSC (Gyr)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

## MCMAC Result

## 1:1 b=0kpc

```python
filename = "../output_default/zh_1to1_b0_z_MCMAC.txt"

df = pd.read_csv(filename, sep='\t')
df = df[['age','sep.kpc','merger', 'delta.t', 'TSC0', 'TSC0.lower', 'TSC0.upper']]

# データの範囲を確認します
print(df['TSC0'].describe())

df.loc[df['merger'] == 'pre', ['TSC0', 'TSC0.lower', 'TSC0.upper']] *= -1

# TSC0の平均と標準偏差を計算します
mean_tsc0 = df['TSC0'].mean()
std_tsc0 = df['TSC0'].std()

# x軸の範囲を設定します
x = np.linspace(df['TSC0'].min(), df['TSC0'].max(), 100)

# 確率密度関数（PDF）を計算します
pdf = norm.pdf(x, mean_tsc0, std_tsc0)

# PDFをプロットします
plt.plot(x, pdf, label='Probability Density Function', color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.xlabel('TSC0')
plt.ylabel('Density')
plt.title('Probability Density Function of TSC0')
plt.legend()
plt.grid(True)
plt.show()
```
